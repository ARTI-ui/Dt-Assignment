```python
import requests
from bs4 import BeautifulSoup
import openpyxl


comp = [
    {"name": "Unilever", "url": "https://www.unilever.com"},
      {"name": "Roche", "url": "https://www.roche.com"},
    {"name": "Nestle Waters", "url": "https://www.nestlewaters.com"},
    {"name": "Sanofi", "url": "https://www.sanofi.com"},
        {"name": "Merck & Co.", "url": "https://www.merck.com"}

]
#Checking the relevance 
def relevant(keywords, text):
    return any(keyword.lower() in text.lower() for keyword in keywords)

def cul_relevance(company_data):
    yes_count = sum(1 for value in company_data.values() if value == "Yes")
    
    # condition to check
    if yes_count >= 7:
        return "Yes (HR)"
    elif yes_count >= 1:
        return "Yes (LR)"
    else:
        return "No"

#Scraping data 
def com_data(company):
    try:
        response = requests.get(company["url"])
        soup = BeautifulSoup(response.text, 'html.parser')
        
        cont = ""
        
        sections_check = ['products', 'about', 'health', 'services', 'nutrition', 'research']
        
        for section in sections_check:
            section_elements = soup.find_all(string=lambda text: text and section.lower() in text.lower())
            for element in section_elements:
                cont += element.get_text()
        
        probiotics_k = ["probiotics", "gut health", "immune system"]
        formulations_k= ["bacipro", "provinorm", "cognisol"]
        categories_k = ["food", "beverages", "pharmaceuticals", "nutrition", "health", "Milk"]
        health_segments_k = {
            "gut_health": ["gut health", "digestive", "microbiome"],
            "women_health": ["women's health", "female health", "pregnancy"],
            "cognitive_health": ["cognitive health", "brain health", "mental clarity"],
            "mental_wellness": ["mental wellness", "mood", "stress"],
            "sports_nutrition": ["sports nutrition", "athletes", "performance"]
        }

        probiotics = relevant(probiotics_k, cont)
        formulations = relevant(formulations_k, cont)
        categories = relevant(categories_k, cont)
        health_segments = {key: relevant(val, cont) for key, val in health_segments_k.items()}

        company_data = {
            "Name": company["name"],
            "URL": company["url"],
            "Probiotics": "Yes" if probiotics else "No",
            "Category": "Yes" if categories else "No",
            "Mfg Type": "Brand" if categories else "Manufacturer",
            "UBL Strains": "Yes" if probiotics else "No",
            "Formulations": "Yes" if formulations else "No",
            "Market Reach": "High",  
            "Region": "Global",  
            "Gut Health": "Yes" if health_segments["gut_health"] else "No",
            "Women’s Health": "Yes" if health_segments["women_health"] else "No",
            "Cognitive Health": "Yes" if health_segments["cognitive_health"] else "No",
            "Mental Wellness": "Yes" if health_segments["mental_wellness"] else "No",
            "Sports Nutrition": "Yes" if health_segments["sports_nutrition"] else "No",
        }

        company_data["Relevance"] = calculate_relevance(company_data)

        return company_data

    except Exception as e:
        print(f"Error scraping data for {company['name']} ({company['url']}): {str(e)}")
        return None

wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Company Data"

headers = [
    "Name", "URL", "Probiotics", "Category", "Mfg Type", "UBL Strains",
    "Formulations", "Market Reach", "Region", "Gut Health", "Women’s Health",
    "Cognitive Health", "Mental Wellness", "Sports Nutrition", "Relevance"
]
ws.append(headers)

for company in comp:
    company_data = com_data(company)
    
    if company_data:
        row = [company_data[key] for key in headers]
        ws.append(row)


wb.save("C:/Users/hp/Desktop/all/practice/data.xlsx")

print("Task2_excel_file.xlsx")
```
